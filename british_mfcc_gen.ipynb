# Notebook to extract MFCC data from British dataset

```python
import librosa
import os
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def mfcc_british_csv_data():
    """
    Only needs to run 1nce to generate the df that has been saved at the bottom
    """
    mfccs_array = np.empty((0, 12), float)
    columns = ['feature_' + str(x) for x in range(1,13,1)]
    word = []
    speaker = []
    
    for filename in os.listdir('intersection_wav-20210409T172655Z-001\\Intersection_wav\\'):

    #    if "a" == filename.split('_')[0]: 

        try:
            y, sr = librosa.load('intersection_wav-20210409T172655Z-001\\Intersection_wav\\' + filename, sr=None)
            mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=12)         
            word.append(filename.split('_')[1])
            speaker.append(filename.split('_')[2].split('.')[0])
            
            mfccs_array = np.append(mfccs_array, np.average(mfccs, axis=1).reshape(1,12), axis=0)
        except:
           print('could not use word {}'.format(filename.split('_')[0]))

    df = pd.DataFrame(mfccs_array, columns=columns)
    df['word'] = word
    df['speaker'] = speaker


    df.to_csv('british_mfcc.csv')
    

mfcc_british_csv_data()
```
